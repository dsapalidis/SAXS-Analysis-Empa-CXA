```python
## Small-Angle X-ray Scattering ##

## .gfrm to .mat Bruker to Matlab Converter (Version 1.2) ##

## Empa, Center for X-ray Analytics, D.Sapalidis, St. Gallen, Switzerland, 12.12.2023 ##

import os
import fabio
import matplotlib.pyplot as plt
from scipy.io import savemat

def process_and_save_images(path):
    # Get names of files without extension
    filenames_without_extension = [os.path.splitext(filename)[0] for filename in os.listdir(path)]

    for filename in os.listdir(path):
        filepath = os.path.join(path, filename)

        try:
            # Run fabio and create the file
            image = fabio.open(filepath)
            header_image = image.header

            # Print image data
            print(image.data)

            # Save image as a MATLAB file
            data_dict = {"image": image.data}
            savemat(os.path.join(path, f"{filename}.mat"), data_dict)
            savemat(os.path.join(path, f"header_{filename}.mat"), header_image)

            # Display the image using matplotlib
            plt.imshow(image.data)
            plt.show()

        except Exception as e:
            print(f"An error occurred: {e}")

# Example usage of the function
path_to_process = 'C:/Users/sapd/Desktop/χχ/'
process_and_save_images(path_to_process)
```
